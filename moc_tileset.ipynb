```python
from PIL import Image

```

```python
image = Image.open('TileFlower.png')
```

```python
tag = "PUG"
color_1 = (110, 10, 10, 255) # background color
color_2 = (5, 50, 5, 255) # flower color
save_name = tag + '_TileFlower.png'
```

```python
new_tileset = Image.new("RGBA", image.size)
```

```python
# for x in range(image.width):
#     for y in range(image.height):
#         r, g, b, a = image.getpixel((x, y))
#         print(r, g, b, a)
```

```python
for x in range(image.width):
    for y in range(image.height):
        r, g, b, a = image.getpixel((x, y))
        if (r, g, b, a) == (6, 26, 31, 255):
            new_tileset.putpixel((x, y), color_1)
        else:
            new_tileset.putpixel((x, y), color_2)   
```

```python
new_tileset.save(save_name)
```
